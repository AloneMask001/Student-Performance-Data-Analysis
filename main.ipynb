# The Rule

## First Case: If the Mean relatively close to Median we called "SYMMETRIC"
## Second Case: If the Mean > Median we called "POSITIVELY SKEWED"
## Third Case: If the Mean < Median we called "NEGATIVELY SKEWED"

* MATHSCORE -> Mean(69.349) ≈ Median(70.000) ---> "SYMMETRIC"
* ReadingScore -> Mean(69.962) ≈ Median(70.000) ---> "SYMMETRIC"
* WritingScore -> Mean(68.601) ≈ Median(70.000) ---> "SYMMETRIC"

## Those operation states that all 3 column approximate symetric.

```python
print(df["Absences"].describe())
print(df["Health"].describe())
df["Absences"] = df["Absences"].fillna(df["Absences"].median())
df["Health"] = df['Health'].fillna(df['Health'].mode()[0])
#We assign absences and health to the median and mode in order to fill nan values.
```

```python
df['FinalGrade'] = ((df['MathScore'] + df['ReadingScore'] + df['WritingScore']) / 3 / 5).round(2)
df['FinalGrade'].unique()
#We calculte the finalgrades again.
```

# TASK 10 — PART 3: Effect of Test Preparation
# ===========================================================

# How did I compare the scores? -> I grouped the dataframe by the 'TestPreparation' column 
  * and calculated the mean for Math, Reading, and Writing scores.
  * To find the exact impact, I applied the percentage improvement formula: ((completed - none) / none) * 100.

# Is there a meaningful difference? -> Yes, the results show a clear and meaningful difference. 
 * Students who completed the test preparation course performed significantly better across all subjects.

# What is the percentage improvement? -> 
 * Math Score improved by 4.446484%
 * Reading Score improved by 15.355923 %
 * Writing Score improved by 11.327265 %

# Conclusion -> The preparation course was highly effective, especially for the Writing score, 
 * which usually shows the highest improvement rate.

```python
# Grouping the data by 'TestPreparation' to get the mean scores
prep_avg = df.groupby('TestPreparation')[['MathScore', 'ReadingScore', 'WritingScore']].mean()
print("--- Average Scores ---")
print(prep_avg)

none_scores = prep_avg.loc['none']
completed_scores = prep_avg.loc['completed']

#Only for this(percentage_improvement) part calculation i ask for help to GOOGLE GEMINI.
percentage_improvement = ((completed_scores - none_scores) / none_scores) * 100
print("\n--- Percentage Improvement (%) ---")
print(percentage_improvement)
```
